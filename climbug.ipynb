```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

BASE_LIST = ["https://cs170.org","https://fa24.datastructur.es","https://cs61a.org","https://www.eecs70.org","https://cs61c.org/fa24","https://fa24.cs161.org","https://people.eecs.berkeley.edu/~sinclair/cs271/f24","https://eecs171.com","https://eecs276.com","https://cs186berkeley.net","https://cs162.org","https://eecs189.org"]
DIR_LIST = ["./cs170","./cs61b","./cs61a","./cs70","./cs61c","./cs161","./cs271","./cs171","./cs276","./cs186","./cs162","./cs189"]
MERGED_BASE_DIR=dict(zip(BASE_LIST,DIR_LIST))

# 创建保存 PDF 的文件夹（如果不存在）
for dir in MERGED_BASE_DIR.values():
    os.makedirs(dir, exist_ok=True)

# 获取网页内容
for url in MERGED_BASE_DIR.keys():
    response = requests.get(url)
    response.raise_for_status()  # 检查请求是否成功
    soup = BeautifulSoup(response.text, "html.parser")

    # 找到所有 PDF 链接
    pdf_links = soup.find_all("a", href=True)
    
    # 下载每一个 PDF
    for link in pdf_links:
        href = link["href"]
        if href.endswith(".pdf"):  # 筛选 PDF 链接
            full_url = urljoin(url, href)  # 拼接完整的 URL
            print(f"正在下载: {full_url}")
            try:
        
                # 下载 PDF 文件
                pdf_response = requests.get(full_url)
                pdf_response.raise_for_status()  # 检查下载是否成功
            except requests.exceptions.HTTPError as e:
                print(f"HTTPError: {e}")
            except requests.exceptions.RequestException as e:
                print(f"请求错误: {e}")
            
            # 获取文件名
            filename = os.path.basename(href)
            filepath = os.path.join(MERGED_BASE_DIR.get(url), filename)
            
            # 保存到本地
            with open(filepath, "wb") as f:
                f.write(pdf_response.content)
            print(f"已保存: {filepath}")
    
    print("所有 PDF 下载完成！")
```
